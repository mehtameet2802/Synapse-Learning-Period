```python
import re 
name = list(train_df.Name)
list2 = []
for i in name:
    search = re.search('(\w+)\.',i)
    list2.append(search.group(1))
train_df['Title'] = list2
train_df
```

```python
dict2 = {}
list2 = []
title_unique = list(train_df.Title.unique())
for j in title_unique:
    count1=0
    total=0
    for i in range(0,891):
        if(train_df.Title[i] == j):
            if(train_df.Survived[i]==1):
                count1+=1
                total+=1
            elif(train_df.Survived[i]==0):
                total+=1
        else:
            continue
        countf = count1/total
        dict2.update({train_df.Title[i]:countf})
    list2.append(countf)
SR_df = pd.DataFrame(list(dict2.items()), columns = ['Title','Survival_rate'])
SR_df
```

```python
y = list2
x = title_unique
fig,ax = plt.subplots(1,1,figsize = (16,5))
ax = plt.bar(x,y)
ax = plt.xlabel("Title")
ax = plt.ylabel("Survival Rate")
```

```python
counts=0
countd=0
countsm=0
countdm=0
countsf=0
countdf=0
for i in range(0,891):
    if(train_df.Survived[i] == 1):
        counts+=1
        if(train_df.Sex[i]=='male'):
            countsm+=1
        else:
            countsf+=1
    else:
        countd+=1
        if(train_df.Sex[i]=='male'):
            countdm+=1
        else:
            countdf+=1
g1 = np.array([counts,countd])
g2 = np.array([countsm,countdm])
g3 = np.array([countsf,countdf])
```

```python
train_df.drop(['PassengerId','Ticket','Name'],axis = 'columns', inplace =True)
train_df
```

```python
nan_dict={}
def isNaN(value):
    try:
        import math
        return math.isnan(float(value))
    except:
        return False

for j in train_df.columns:
    count=0
    total=0
    for i in range(0,891):
        if(isNaN(train_df[j][i])):
            count+=1
            total+=1
        else:
            total+=1
    percent = (count/total)*100
    nan_dict.update({j:percent})
```

```python
l = len(nan_dict)
keys = list(nan_dict.keys())
value = list(nan_dict.values())
for i in range(0,l):
    if (value[i]>50):
        for j in range(0,l-1):
            if(keys[i]==train_df.columns[j]):
                train_df.drop(train_df.columns[j], axis = 1,inplace =True)
```

```python
mode = train_df.Embarked.mode()
train_df.Embarked.fillna(f"{mode}", inplace = True)
```

```python
mean = train_df.Age.mean()
train_df.Age.fillna(mean, inplace = True)
```

```python
nan_dict2 = {}
for j in train_df.columns:
    count=0
    total=0
    for i in range(0,891):
        if(isNaN(train_df[j][i])):
            count+=1
            total+=1
        else:
            total+=1
    percent = (count/total)*100
    nan_dict2.update({j:percent})
```

```python
x = list(nan_dict2.values())
y = list(nan_dict2.keys())
fig,ax = plt.subplots(1, 1, figsize = (12,4))
ax.bar(y,x, width = 0.5)
```

```python
fm_list =[]
for i in range(0,891):
    fm_list.append(train_df.SibSp[i]+train_df.Parch[i])
train_df['Family_members'] = fm_list
```

```python
train_df.drop(['Parch','SibSp'], inplace=True, axis=1)
```

```python
train_df.Age = (train_df.Age-train_df.Age.min())/(train_df.Age.max()-train_df.Age.min())
train_df.Fare = (train_df.Age-train_df.Fare.min())/(train_df.Fare.max()-train_df.Fare.min())
train_df.Family_members = (train_df.Age-train_df.Family_members.min())/(train_df.Family_members.max()-train_df.Family_members.min())
```

```python
dummy_Pclass = pd.get_dummies(train_df.Pclass)
dummy_Sex = pd.get_dummies(train_df.Sex)
dummy_Title = pd.get_dummies(train_df.Title)
dummy_Embarked = pd.get_dummies(train_df.Embarked)
```

```python
train_df = pd.merge(left=train_df,right=dummy_Pclass,left_index=True,right_index=True)
train_df = pd.merge(left=train_df,right=dummy_Sex,left_index=True,right_index=True)
train_df = pd.merge(left=train_df,right=dummy_Title,left_index=True,right_index=True)
train_df = pd.merge(left=train_df,right=dummy_Embarked,left_index=True,right_index=True)
train_df
```
