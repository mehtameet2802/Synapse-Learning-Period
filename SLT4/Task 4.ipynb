# Importing all requirements

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation,Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
from sklearn.model_selection import train_test_split
```

```python
import numpy as np
import pandas as pd
```

# Create a model

```python
#create a model using sequential
#add dense layers and activation function
```

```python
model = Sequential([
    Dense(units = 32, input_dim =16, activation = 'relu'),
    Dense(units = 16, activation = 'relu'),
    Dense(units = 1, activation = 'sigmoid')
])
```

# Importing Dataset

```python
#import your dataset
train_df = pd.read_csv('train2.csv')
test_df = pd.read_csv('test.csv')
```

# Create features and run code

```python
#create the features and labels of the dataset
#run the code
features = ["Age","Fare","Family_members","1","2","3","male","female","C","Q","S","Master","Miss","Mr","Mrs","Ms"]
X = train_df[features]
Y = train_df["Survived"]
```

# Split the dataset

```python
#Split the dataset into train,test dataset
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.20,random_state = 1)
```

# Compile the dataset

```python
#compile your model using optimizers like "SGD","ADAM"
#use model.fit to train 
model.compile(optimizer = "SGD", loss = 'binary_crossentropy', metrics = ["binary_accuracy"])
model_result = model.fit(X,Y, batch_size = 100, epochs = 200, validation_split = 0.2, shuffle = True, verbose = 2)
```

# Prediction

```python
#use model.predict to predict the survival rate on the test dataset and print the predictions
#Try to understand what does these numbers in predict means
prediction = model.predict(X_test)
prediction
```
