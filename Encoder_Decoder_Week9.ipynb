```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
```

```python
header_list = ["English","Russian","3"]
```

```python
df_data = pd.read_table("rus.txt",names = header_list)
```

```python
df_data = df_data.drop("3",axis = "columns")
```

```python
df_data
```

```python
X = df_data["English"]
y = df_data["Russian"]
X = X.head(10000)
y = y.head(10000)
y = '\t'+y+'\n'
```

```python
y
```

```python
input_texts = []
for i in X:
    input_texts.append(i)
output_texts = []
for i in y:
    output_texts.append(i)
```

```python
input_texts
```

```python
import tensorflow as tf
unique_l1 = []
for i in X:
    for j in i:
        for k in j:
            if(j not in unique_l1):
                unique_l1.append(k)

#for i in X:
#    for j in i.split():
#        if(j not in unique_l1):
#            unique_l1.append(j)

unique_l1 = sorted(list(unique_l1))
unique_l1
```

```python
unique_l2 = []
for i in y:
    for j in i:
        for k in j:
            if(k not in unique_l2):
                unique_l2.append(k)
unique_l2 = sorted(list(unique_l2))
unique_l2
```

```python
print(len(unique_l1),len(unique_l2))
```

```python
#inputs and outputs for decoder
num_input = len(X)
num_encoder_tokens = len(unique_l1)
max_encoder_tlen = X.str.len().max()
num_output = len(y)
num_decoder_tokens = len(unique_l2)
max_decoder_tlen = y.str.len().max()
```

```python
input_token_index = {}
target_token_index = {}
for i in range(num_encoder_tokens):
    input_token_index[unique_l1[i]] = i
for i in range(num_decoder_tokens):
    target_token_index[unique_l2[i]] = i
print(input_token_index)
print(target_token_index)
```

```python
#converting to one hot encoder
encoder_input_data = np.zeros((len(input_texts), max_encoder_tlen, num_encoder_tokens), dtype="float32"
)
decoder_input_data = np.zeros((len(input_texts), max_decoder_tlen, num_decoder_tokens), dtype="float32"
)
decoder_target_data = np.zeros((len(input_texts), max_decoder_tlen, num_decoder_tokens), dtype="float32"
)

for i, (input_text,target_text) in enumerate(zip(input_texts,output_texts)):
    for t, char in enumerate(input_text):
        encoder_input_data[i,t,input_token_index[char]] = 1.0
    encoder_input_data[i,t+1:,input_token_index[' ']] = 1.0
    for t, char in enumerate(target_text):
        #decoder_target_data is ahead of decoder_input_databy one timestep
        decoder_input_data[i, t, target_token_index[char]] = 1.0
        if t>0:
            #decoder_output will be ahead by one timestep
            #and will not include the start character
            decoder_target_data[i,t-1,target_token_index[char]] = 1.0
    decoder_input_data[i,t+1:,target_token_index[' ']] = 1.0
    decoder_target_data[i,t:,target_token_index[' ']] = 1.0
```

```python
print(encoder_input_data[0].shape)
print(decoder_input_data.shape)
print(decoder_target_data.shape)
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense
```

```python
#encoder model
latent_dim = 256
encoder_inputs = Input(shape = (None,num_encoder_tokens))
encoder = LSTM(latent_dim,return_state = True)
encoder_outputs,state_h,state_c = encoder(encoder_inputs)
#discarding the encoder output and only considering hidden and inital state
encoder_states = [state_h,state_c]


#decoder model
decoder_inputs = Input(shape=(None,num_decoder_tokens))
decoder_lstm = LSTM(latent_dim,return_sequences = True, return_state = True)
decoder_outputs, _,_ = decoder_lstm(decoder_inputs,initial_state = encoder_states)
decoder_dense = Dense(num_decoder_tokens, activation = 'softmax')
decoder_outputs = decoder_dense(decoder_outputs)
```

```python
model = Model([encoder_inputs,decoder_inputs],decoder_outputs)

model.compile(optimizer = 'rmsprop', loss = 'categorical_crossentropy', metrics = ['accuracy'])
model.save("s2s")
```

```python
#model.summary()
```

```python
model.fit([encoder_input_data,decoder_input_data],decoder_target_data,batch_size=125,epochs = 100,validation_split=0.2)
```

```python
encoder_model = Model(encoder_inputs, encoder_states)

decoder_state_input_h = keras.Input(shape=(latent_dim,))
decoder_state_input_c = keras.Input(shape=(latent_dim,))
decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]
decoder_outputs, state_h, state_c = decoder_lstm(
   decoder_inputs, initial_state=decoder_states_inputs
)
decoder_states = [state_h, state_c]
decoder_outputs = decoder_dense(decoder_outputs)
decoder_model = Model(
   [decoder_inputs] + decoder_states_inputs, [decoder_outputs] + decoder_states
)
```

```python
# model = keras.models.load_model("s2s")

# encoder_inputs = model.input[0]
# encoder_outputs,state_h_e,state_c_e = model.layers[2].output
# encoder_states = [state_h_e,state_c_e]

# encoder_model = keras.Model(encoder_inputs,encoder_states)

# decoder_inputs = model.input[1]

# decoder_state_input_h = Input(shape=(latent_dim,))
# decoder_state_input_c = Input(shape=(latent_dim,))
# decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]

# decoder_lstm = model.layers[3]

# decoder_outputs, state_h, state_c = decoder_lstm(decoder_inputs, initial_state=decoder_states_inputs)

# decoder_dense = model.layers[4]

# decoder_states = [state_h, state_c]
# decoder_outputs = decoder_dense(decoder_outputs)

# decoder_model = keras.Model([decoder_inputs] + decoder_states_inputs,[decoder_outputs] + decoder_states)

reverse_input_char_index = dict(
    (i,char) for char,i in input_token_index.items())
reverse_target_char_index = dict(
    (i,char) for char,i in target_token_index.items())


def decode_sequence(input_seq):
    # Encode the input as state vectors.
    states_value = encoder_model.predict(input_seq)

    # Generate empty target sequence of length 1.
    target_seq = np.zeros((1, 1,num_decoder_tokens))
    # Populate the first character of target sequence with the start character.
    target_seq[0, 0, target_token_index['\t']] = 1.0
    #target_seq[0, 0] =  runique_dict['\t']

    # Sampling loop for a batch of sequences
    # (to simplify, here we assume a batch of size 1).
    stop_condition = False
    decoded_sentence = ''
    while not stop_condition:
        output_tokens, h, c = decoder_model.predict(
            [target_seq] + states_value)

        # Sample a token
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_char = reverse_target_char_index[sampled_token_index]
        decoded_sentence += sampled_char

        # Exit condition: either hit max length
        # or find stop character.
        if (sampled_char == '\n' or
           len(decoded_sentence) > max_decoder_tlen ):
            stop_condition = True

        # Update the target sequence (of length 1).
        target_seq = np.zeros((1, 1,num_decoder_tokens))
        target_seq[0, 0, sampled_token_index] = 1.

        
        # Update states
        states_value = [h, c]

    return decoded_sentence
```

```python
for s_index in range(100):
    input_s = encoder_input_data[s_index:s_index+1]
    decoded_sent = decode_sequence(input_s)
    print('Input sentence', X[s_index])
    print('Decoded sentence', decoded_sent)
    print('------')
```

```python
# For Machine Translation encoder decoder model is used as instructed.
# Using this model input is taken in English and converted to Russian language. Before passing the input to encoder 
# the imported data is divided into 2 lists of English and Russian data
# preprocessing is carried out on the data in which /t and /n is added to start and end of Russian data respectively so that it is known when the 
# sentence starts and ends.
# English and Russian data is converted in form of list and then a dictionary is made out of the list so it is easy to form a 3D vector of the input data.
# After this the English data is feed into the encoder which consists of LSTM layer, the output of the encoder is not taken into consideration,
# only the hiden states of encoder are taken into consideration, using the hidden states a context vector is formed and then passed to the decoder.
# Decoder has been trained to predict the output based on the previous inputs so using the context vector decoder tries to predict the 
# Russian translation of respective English word.
# The model has been trained on 100 epochs to increase the accuracy of the outputs.
# The hidden states of each decoder is passed on to the next decoder which helps in the prediction of the next word and when 
# the end is reached decoder stops.
# At each and every step of decoder the output is appended to form the final sentence and when the sentence is over
# final output is displayed in Russian
```
